I. Importing the Data. Import a music library .csv file using pandas

```python
#Import pandas
import pandas as pd

#Read the .csv in a dataframe (df)

df = pd.read_csv("Strings Library - Ensemble Music.csv")

#Display partial .csv file to check if it worked. 
df


```

II. Cleaning and sorting the data. Cleaning up the data and filling in gaps.

```python
#Replace empty cells in the Instrumentation column with "String Orchestra" (default)
df['Instrumentation'] = df['Instrumentation'].fillna('String Orchestra')

#Replace empty cells in the Collection column with "none".
df['Collection'] = df['Collection'].fillna('none')

#Replace empty cells in the Performance History column with "none".
df['Performance History'] = df['Performance History'].fillna('none')

#Delete the last Unnamed column.
df.drop(columns='Unnamed: 10', inplace = True)

#Check to see if it worked.
df
```

Some more clean up.

```python
#Replace empty cells in the Collection column with "none".
df['Theme'] = df['Theme'].fillna('none')
```

```python
#Replace empty cells in the Collection column with "none".
df['Notes'] = df['Notes'].fillna('none')

#Replace empty cells in the Arranger, Composer First and Last Names columns with "unlisted".
df['Composer First Name'] = df['Composer First Name'].fillna('unlisted')

#Replace empty cells in the Composer First and Last Names columns with "unlisted".
df['Composer Last Name'] = df['Composer Last Name'].fillna('unlisted')

#Replace empty cells in the Composer First and Last Names columns with "unlisted".
df['Arranger'] = df['Arranger'].fillna('unlisted')

#Replace missing values in the Grade column with "2".
df['Arranger'] = df['Arranger'].fillna('2')

#Check
df
```

```python
#Sort the df by Grade and Title

df = df.sort_values(['Grade', 'Title'])

df
```

```python
#Convert Grade column values to float values

df['Grade'] = df['Grade'].astype(float)

df
```
